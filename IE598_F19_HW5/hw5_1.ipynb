```python
import pandas as pd
import sys
import matplotlib.pyplot as plt
import seaborn as sns

# Read in file
df = pd.read_csv('hw5_treasury yield curve data.csv')

# Get the row and column number
ROW = df.shape[0]
COL = df.shape[1]
print("Number of Rows of Data = " + str(ROW))
print("Number of Columns of Data = " + str(COL) + '\n')

df.head()
```

```python
# Remove missing values
df=df.dropna()

print("Data types in 'df': ")
print(df.dtypes, "\n")

print("Missing data in 'df': ")
print(df.isnull().sum(), "\n")

# Displays the number of times each specific value of 'MEDV' in 'df'
df['Adj_Close'].value_counts()
```

```python
# Show graphical summary of the relationships in 'df'
cols = ['SVENF01', 'SVENF07', 'SVENF13', 'SVENF24', 'SVENF30']
sns.pairplot(df[cols], size=2.5)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Respective linear correlations table
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1)
hm = sns.heatmap(cm,cbar=True,annot=True,square=True,fmt='.2f',annot_kws={'size': 15},yticklabels=cols, xticklabels=cols)           
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

_ = sns.swarmplot(x='SVENF07', y='Adj_Close', data=df)
# Label the axes
_ = plt.xlabel('SVENF07')
_ = plt.ylabel('Adj_Close')
# Show the plot
plt.show()

_ = sns.swarmplot(x='SVENF24', y='Adj_Close', data=df)
# Label the axes
_ = plt.xlabel('SVENF24')
_ = plt.ylabel('Adj_Close')
# Show the plot
plt.show()
```

```python
# covariance for SVENF07 and SVENF24
cov_matrix = np.cov(df['SVENF07'], df['Adj_Close'])
print('covariance between SVENF07 and Adj_Close: %.4f' % (cov_matrix[0,1]))

cov_matrix = np.cov(df['SVENF24'], df['Adj_Close'])
print('\ncovariance between SVENF24 and Adj_Close: %.4f' % (cov_matrix[0,1]))
```
